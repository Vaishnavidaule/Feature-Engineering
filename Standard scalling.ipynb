### Label Encoding  
The area variable is originally in categorical datatype, so we converted it into numerical values to improve accuracy

```python
from sklearn.preprocessing import LabelEncoder

# Load the data into a DataFrame
df1 = pd.DataFrame(df) 

# Create a LabelEncoder object
label_encoder = LabelEncoder()

# Encode the "Area" column
df1['Area_encoded'] = label_encoder.fit_transform(df1['Area'])

# Display the DataFrame with the encoded "Area" column
df1

```

```python
df1 = df1.drop('Area', axis=1)  # Specify axis=1 to indicate dropping a column, not a row
df1
```

```python
df1.dropna(inplace=True)
```

```python
df1.reset_index(drop=True, inplace=True)
```

```python
df1.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df1.drop('Profit',axis =1),df1['Profit'],test_size=0.20,random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
ss.fit(X_train)
```

```python
ss.var_
```

```python
ss.mean_
```

```python
X_train.describe()
```

```python
# here we give both train and test data that why we predict the data
X_train_scaled = ss.transform(X_train)
X_test_scaled = ss.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled,columns= X_train.columns)
```

```python
# scatter plot used here
fig,(ax1,ax2)= plt.subplots(ncols = 2, figsize = (12,5))

ax1.scatter(X_train["Marketing Spend"],X_train["Administration"],color = "red")
ax1.set_title("Before Scaling")

ax2.scatter(X_train_scaled["Marketing Spend"],X_train_scaled["Administration"],color = "Yellow")
ax2.set_title("After Scalling")
```

```python
# KDE use here
# ax = axis 
#ax1 and ax2 = is a figure
fig,(ax1,ax2)= plt.subplots(ncols = 2, figsize = (12,5))

ax1.set_title("Before Scalling")
sns.kdeplot(X_train["Marketing Spend"],ax=ax1)
sns.kdeplot(X_train["Administration"],ax=ax1)

ax2.set_title("After Scalling")
sns.kdeplot(X_train_scaled["Marketing Spend"],ax=ax2)
sns.kdeplot(X_train_scaled["Administration"],ax=ax2)

```

```python
# scatter plot used here
fig,(ax1,ax2)= plt.subplots(ncols = 2, figsize = (12,5))

ax1.scatter(X_train["Transport"],X_train["Area_encoded"],color = "red")
ax1.set_title("Before Scaling")

ax2.scatter(X_train_scaled["Transport"],X_train_scaled["Area_encoded"],color = "Yellow")
ax2.set_title("After Scalling")
```

```python
# KDE use here
# ax = axis 
#ax1 and ax2 = is a figure
fig,(ax1,ax2)= plt.subplots(ncols = 2, figsize = (12,5))

ax1.set_title("Before Scalling")
sns.kdeplot(X_train["Transport"],ax=ax1)
sns.kdeplot(X_train["Area_encoded"],ax=ax1)

ax2.set_title("After Scalling")
sns.kdeplot(X_train_scaled["Transport"],ax=ax2)
sns.kdeplot(X_train_scaled["Area_encoded"],ax=ax2)

```
