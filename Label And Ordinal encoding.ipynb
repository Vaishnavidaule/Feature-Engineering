# Split the data in train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[["review", "education"]],
                                                    df["purchased"],
                                                    test_size=0.20,
                                                    random_state=42)
```

## insights
* here we split the data into train and test 
* in X_train we take 2 column that is review and education and in y_train we take purchased column
* Test data size is specified as 20% of the dataset using the test_size=0.2 parameter.
